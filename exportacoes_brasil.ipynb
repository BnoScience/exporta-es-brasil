```python
def column_to_list(data, index):

    """ Retorna uma lista com os valores de uma coluna, para cada linha no dataset.
    Argumentos:
    data -- dataset
    index -- valor para encontrar os valores da coluna do dataset
    """
    column_list = []
    for i in range(len(data)):
        column_values = data[i][index]
        column_list.append(column_values)
    return column_list
```

```python
# Tabelas auxiliares

# Tabela do países
paises = pd.read_csv('paises.csv', encoding='latin1', sep=';', usecols=['CO_PAIS', 'NO_PAIS'])

# Tabela dos produtos exportados
produtos = pd.read_csv('grupos_produtos.csv', encoding='latin1', sep=';', usecols=['CO_NCM', 'CO_EXP_SUBSET', 'NO_EXP_SUBSET_POR'])

# Tabela das vias
vias = pd.read_csv('vias.csv', encoding='latin1', sep=';')

# Tabela do CO_URF
co_urf = pd.read_csv('co_urf.csv', encoding='latin1', sep=';')

# Medidas
unidades_medida = pd.read_csv('unidades.csv', encoding='latin1', sep=';')
```

```python
tabelas = [(paises, 'paises'), (produtos,'produtos'),  
           (vias, 'vias'), (co_urf,'co_urf') , (unidades_medida, 'unidades_medida')]

# Vendo o tamanho das tabelas
for tabela, nome in tabelas:
    linhas, colunas = tabela.shape    
    print("-> Tabela: {}, linhas: {}, colunas {}\n ".format(nome, linhas, colunas))
```

```python
# Remover após a junção dos dados
colunas_remover = ['CO_NCM', 'CO_UNID', 'CO_PAIS', 'CO_VIA', 'CO_URF', 'CO_EXP_SUBSET']

# Incluindo o país
df_final = pd.merge(df, paises, on='CO_PAIS', how='inner')

# Incluindo produtos
df_final = pd.merge(df_final, produtos, on='CO_NCM', how='inner')

# Incluindo as vias
df_final = pd.merge(df_final, vias, on='CO_VIA', how='inner')

# Incluindo CO_URF
df_final = pd.merge(df_final, co_urf, on='CO_URF', how='inner')

# Incluindo as medidas
df_final = pd.merge(df_final, unidades_medida[['CO_UNID', 'SG_UNID']], on='CO_UNID', how='inner')

# Excluindo as colunas desnecessárias
df_final = df_final.drop(axis=1, columns=colunas_remover)
```

```python
df_final = df_final.drop_duplicates()
```

```python
df_final.isnull().sum()
```

```python
# Convertendo os anos em strings e extraindo o primeiro item
ano = df_final.CO_ANO.astype('str')
ano = [ano.split('.')[0] for ano in ano]

# Convertendo os meses em strings e extraindo o primeiro item
mes = df_final.CO_MES.astype('str')
mes = [mes.split('.')[0] for mes in mes]

# Concatenando mes-ano
mes_ano = pd.Series(mes)+'-'+pd.Series(ano)

# Convertendo em datetime
mes_ano = pd.to_datetime(mes_ano, infer_datetime_format=True)

# Salvando como index do dataset
df_final = df_final.set_index(mes_ano)

# Confirmando se os dados estão corretos
df_final[['CO_MES', 'CO_ANO']].head()
```
